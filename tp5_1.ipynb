##### Ajuste lineal

```python
import numpy as np
x = np.array([7.4, 7.5, 8.5, 9.0, 9.0, 11.0, 13.0, 14.0, 14.5, 16, 17.0, 18.0, 20.0, 23.0])
y = np.array([30.0, 25.0, 31.5, 27.5, 39.5, 38.0, 43.0, 49.0, 55.0, 48.5, 51.0, 64.5, 63.0, 68.0])
b = np.cov(x, y, bias=True)[0, 1]/np.var(x)
a = np.mean(y)-b*np.mean(x)

import matplotlib.pyplot as plt
fig, ax = plt.subplots()
plt.title('Gráfico de dispersión')
ax.scatter(x, y)
ye = b*x+a
ax.plot(x, ye)
plt.show()
r2 = sum((ye-np.mean(y))**2)/sum((y-np.mean(y))**2)
r2
```

##### Ajuste parabólico

```python
x2 = x ** 2
x3 = x ** 3
x4 = x ** 4

n = len(x)
m = np.array([[n, sum(x), sum(x2)], [sum(x), sum(x2), sum(x3)], [sum(x2), sum(x3), sum(x4)]])
b = np.array([sum(y), sum(x*y), sum(x2*y)])

s=np.linalg.solve(m, b)
print(m, b, s)
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
plt.title('Gráfico de dispersión')
ax.scatter(x, y)
ye = s[0]+s[1]*x+s[2]*x**2
ax.plot(x, ye)
plt.show()
```

```python
r2 = sum((ye-np.mean(y))**2)/sum((y-np.mean(y))**2)
r2
```
