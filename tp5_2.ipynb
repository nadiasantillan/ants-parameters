##### Ajuste lineal

```python
import numpy as np
x = np.array([0.31, 0.31, 0.56, 0.56, 0.89, 0.89, 0.96, 0.96, 1.15, 1.15, 1.25])
y = np.array([0.2, 0.1, 0.0, 0.2, 0.3, 0.5, 0.0, 0.6, 0.4, 0.2, 0.5])

b=np.cov(x, y, bias=True)[0, 1]/np.var(x)
a=np.mean(y)-b*np.mean(x)
ye = b*x+a

import matplotlib.pyplot as plt
fig, ax = plt.subplots()
plt.title('Gráfico de dispersión')
ax.scatter(x, y)
ax.plot(x, ye)
plt.show()
```

Coeficientes de correlación Y/X y X/Y

```python
rxy = np.cov(x, y, bias=True)/(np.std(x)*np.std(y))
ryx = np.cov(y, x, bias=True)/(np.std(x)*np.std(y))

print("Y/X", rxy[0,1])
print("X/Y", ryx[0,1])
```

Coeficiente de determinación

```python
r2 = sum((ye-np.mean(y))**2)/sum((y-np.mean(y))**2)
print("R2", r2)
```

Ajuste parabólico

```python
x2 = x ** 2
x3 = x ** 3
x4 = x ** 4

n = len(x)
m = np.array([[n, sum(x), sum(x2)], [sum(x), sum(x2), sum(x3)], [sum(x2), sum(x3), sum(x4)]])
b = np.array([sum(y), sum(x*y), sum(x2*y)])

s=np.linalg.solve(m, b)
print(m, b, s)

ye = s[0]+s[1]*x+s[2]*x2

import matplotlib.pyplot as plt
fig, ax = plt.subplots()
plt.title('Ajuste parabólico')
ax.scatter(x, y)
ax.plot(x, ye)
plt.show()
```

Coeficiente de determinación

```python
r2 = sum((ye-np.mean(y))**2)/sum((y-np.mean(y))**2)
print("R2", r2)
```
